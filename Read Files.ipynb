```python
# Dealing with the newline \n
# Reading a String and Stripping the Newline from it

def read_line():
    # Open file : friends.txt
    onfile = open('friends.txt','r')
    
    # Read File Contents
    line1=onfile.readline()
    line2=onfile.readline()
    line3=onfile.readline()
    
    #Rstrip the \n from each string
    line1=line1.rstrip('\n')
    line2=line2.rstrip('\n')
    line3=line3.rstrip('\n')
    
    # close file
    onfile.close()
    
    # print data
    print(line1)
    print(line2)
    print(line3)
    
# call 
read_line()
```

```python
#########The user entered sales data for five days as follows: 225, 255, 240, 135, 560.
def main():
    # Oppen File
    outfile = open('sales.txt', 'w')

    # Enter Number oof dayes
    num_days = int(input('Enter the number of days of sales: '))

    # Looppp and add values
    for day in range(1, num_days + 1):
        sales = float(input(f'Enter sales for day {day}: '))
        outfile.write(str(sales) + '\n')

    # Close
    outfile.close()
    print('Sales amounts have been written to sales.txt.')
# Call Function
main()
```

```python
#Processing Records

def main():
    emp_file = open('employees.txt', 'r')

    name = emp_file.readline()

    while name != '':
        id_num = emp_file.readline()
        dept = emp_file.readline()   

        name = name.rstrip('\n')
        id_num = id_num.rstrip('\n')
        dept = dept.rstrip('\n')

        print('Name:', name)
        print('ID:', id_num)
        print('Dept:', dept)
        print()  
        name = emp_file.readline()

    emp_file.close()

main()
```

```python
# Open File Not Found
def main():
    try:
        emp_file = open('non_existing_file.txt', 'r')
        print('File opened successfully!')        
        content = emp_file.read()
        print(content)     
        emp_file.close()
    except FileNotFoundError:
        print('Error: The file does not exist.')

    except Exception as e:
        print(f'An unexpected error occurred: {e}')

main()
```
